```python
alpha = np.array([5442,5484])
canals = np.array([6850,6902])

reg = linear_model.LinearRegression()
newcanals = canals.reshape(-1,1)
reg.fit(newcanals,alpha)
m = reg.coef_
b = reg.intercept_

plt.scatter(canals,alpha, label = r"Kalibrationspunkte",color="orange")
plt.plot(canals, m*canals+b, label = "Linearer Fit: ("+ str(np.round(float(m),2))+"*Kanal"+str(np.round(b,2))+")eV")
plt.xlabel("Kanal")
plt.ylabel("Energie in eV")
plt.legend(loc = 0)
plt.savefig("Plots/linearfit.pdf")
```

```python
sums, maxlist = [],[]
for i in range(23):
    valpermin = values[i]/times[i]
    maximum = np.argmax(valpermin[200:])
    maxlist.append(maximum)
    upper = maximum+200
    lower = maximum-200
    sums.append(sum(valpermin[lower:upper]))
plt.plot(rangevalues[0:23],sums,label= r"Maximum $\pm$200 Kanäle")
plt.plot(rangevalues[0:23],maxlist,label = r"Maximum")
plt.legend(loc=0)
plt.xlabel("Druck in mbar")
plt.title("Integration über Maximum/Intervall")
plt.ylabel(r"$\alpha$-Teilchen/Minute")
plt.savefig("Plots/Teilchen je Minute.pdf")
```
