```python
#Berechung der Maxima 
sums2, maxlist = [],[]
for i in range(21):
    valpermin2 = values[i+1]/times[i+1]
    maximum2 = np.argmax(valpermin2[200:])
    maxlist.append(maximum2)
    upper2 = maximum2+200
    lower2 = maximum2-200
    sums2.append(sum(valpermin2[lower2:upper2]))

#Druck -> Schichtdicke
m3 = 3/1000
thickness = m3*rangevalues[1:21]

#Lineare regression
reg2 = linear_model.LinearRegression()
newval = np.array(thickness).reshape(-1,1)
reg2.fit(newval,sums2[1:21])
m2 ,b2 = reg2.coef_, reg2.intercept_

#Bestimmung extrapolierte Reichweite
xt = float(-b2/m2)
print(xt)

xax = np.arange(0,11,1)

plt.plot(thickness,sums2[1:21],label= r"Maximum $\pm$200 Kan√§le")
plt.errorbar(thickness,sums2[1:21],fmt=".",markersize="1",xerr=(xerr[1:21])*m3,capsize=2,label = "x-Error")
plt.plot(xax,m2*xax+b2, label="Linearer Fit, m="+str(np.round(float(m2),2))+", b="+str(np.round(b2,2)))
plt.legend(loc=0)
plt.xlabel("Schichtdicke [cm]")
plt.axis([0,10,0,6250])
plt.annotate("xt="+str(np.round(xt,2))+"cm",xy=(9.5,0),xytext = (8.4,1000),arrowprops = dict(arrowstyle = "->"),bbox = dict(boxstyle ="round", fc ="0.8"))
plt.title("Integration √ºber Maximum/Intervall")
plt.ylabel(r"$\alpha$-Teilchen/Minute")
plt.savefig("Plots/transmissionlinearfit.pdf")
```

```python
sums2, maxlist = [],[]
for i in range(21):
    valpermin2 = values[i+1]/times[i+1]
    maximum2 = np.argmax(valpermin2[200:])
    maxlist.append(maximum2)
    upper2 = maximum2+200
    lower2 = maximum2-200
    sums2.append(sum(valpermin2[lower2:upper2]))

#Druck -> Schichtdicke
m3 = 3/1000
thickness = m3*rangevalues[1:21]

#Anzahl AlphaTeilchen -> Warscheinlichkeit des Auftreffens
upperlimit = np.max(sums2[1:21])
reg2 = linear_model.LinearRegression()
newval = np.array(thickness).reshape(-1,1)
reg2.fit(newval,sums2[1:21])
m2 ,b2 = reg2.coef_/upperlimit, reg2.intercept_/upperlimit

#Halbwertschichtdicke
x1h = (0.5 - b2)/(m2)

#extrapolierte Reichweite
xt = float(-b2/m2)

midline = [0,4.58]
midy = [0.5,0.5]

middown = [4.58,4.58]
middowny = [0,0.5]


plt.plot(thickness,sums2[1:21]/upperlimit,label= r"Maximum $\pm$200 Kan√§le")
plt.errorbar(thickness,sums2[1:21]/upperlimit,xerr=(xerr[1:21])*m3,label = "x-Error",fmt=".",markersize="2",capsize=2)
plt.plot(xax,(m2*xax+b2), label="Linearer Fit, m="+str(np.round(float(m2),2))+", b="+str(np.round(b2,2)))
plt.plot(midline,midy, color = "grey", linestyle = "--")
plt.plot(middown,middowny,color = "grey", linestyle = "--")
plt.legend(loc=0)
plt.axis([0,10,0,1.0])
plt.annotate(r"$x_\mathrm{1h}$="+str(float(np.round(x1h,2)))+"cm",xy=(4.58,0),xytext = (5,0.2),arrowprops = dict(arrowstyle = "->"),bbox = dict(boxstyle ="round", fc ="0.8"))
plt.annotate(r"$x_\mathrm{t}$="+str(np.round(xt,2))+"cm",xy=(9.5,0),xytext = (8.3,0.3),arrowprops = dict(arrowstyle = "->"),bbox = dict(boxstyle ="round", fc ="0.8"))
plt.xlabel("Schichtdicke [cm]")
plt.title("Verteilungsfunktion")
plt.ylabel(r"Warscheinlichkeit W")
plt.savefig("Plots/probabilitycurve.pdf")
```

```python
#Abschw√§chungskoeffizient
n = (np.pi)/2 * (x1h/(xt-x1h))**2
my = (n-(1/3))/x1h

#Mittlere Reichweite
lam = x1h/(n-(1/3))
xmittel = n * lam

print("Schw√§chungskoeffizient= "+str(float(np.round(my,2))),";", "Mittlere Reichweite = "+str(float(np.round(xmittel,2))))
```

```python
#Bragg Kruve Berechnung
fullthickness = m3*rangevalues[0:23]
t3 = m3*rangevalues[0:22]
de, dx,braggy = [],[],[]

maxlist3 = []
for i in range(23):
    valpermin3 = values[i]/times[i]
    maximum3 = np.argmax(valpermin3[200:])
    maxlist3.append(maximum3)

#dE und dx berechnen
for k in range(len(maxlist3)-1):
    de.append(maxlist3[0] - maxlist3[k+1])
    dx.append(fullthickness[k+1])

#dE/dx
for i in range(len(de)):
    braggy.append(de[i]/dx[i]*1/1000)

print(len(braggy))
plt.plot(t3,braggy,label = "Bragg-Kurve")
plt.errorbar(t3,braggy,xerr=(xerr[0:22])*m3,label = "x-Error",fmt=".",markersize="2",capsize=2)
plt.legend(loc=0)
plt.xlabel("Wegl√§nge [cm]")
plt.ylabel("Energieverlust [MeV/cm]")
plt.title("Bragg Kurve")
plt.savefig("Plots/Braggcurve.pdf")
print(maxlist3)
```
